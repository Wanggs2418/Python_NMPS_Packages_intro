```python
a = np.arange(0.0, 5.0, 0.02)
plt.plot(a, np.cos(2*np.pi*a), 'r--')
plt.xlabel("时间/t",fontproperties='Kaiti', fontsize=12,  color='black')
plt.ylabel("振幅/s",fontproperties='Kaiti', fontsize=12)
plt.title(r'余弦波实例 $y=\cos(\frac{2}{\pi} x)$',fontproperties='Kaiti', fontsize=16)
plt.text(1,1, r'$\mu=10$')

# shrink:表示箭头与文本和指向的地方有一定的距离 
arrow_dict = {
    'facecolor':'black',
    'shrink':0.1,
    'width':2
}
plt.annotate(r'$\mu=100$', xy=(2, 1), xytext=(3, 1.5), arrowprops=arrow_dict)

# 设置坐标轴尺度
plt.axis([-1, 6, -2, 2])
plt.grid(True)
plt.show()
```

网格类 GridSpec 的使用

GridSpec + subplot == subplot2grid

```python
import matplotlib.gridspec as gridspec
# 设计网格
gs = gridspec.GridSpec(3, 3)
ax1 = plt.subplot(gs[0, :])
ax2 = plt.subplot(gs[1, :-1])
ax3 = plt.subplot(gs[1:, -1])
ax4 = plt.subplot(gs[2, 0])
ax5 = plt.subplot(gs[2, 1])
```

### 函数图标的绘制

极坐标图的绘制

采用面向对象绘制

<img src="img/10.jpg" style="zoom:50%;" />


```python
# 极坐标图的绘制
np.random.seed(1)
N = 20 #数据的个数
theta = np.linspace(0.0, 2*np.pi, N, endpoint=False)
# 创造 [0, 1) 的均匀分布
raddi = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.randn(N)

ax = plt.subplot(111, projection='polar')
# 对应 left,height,width
bars = ax.bar(theta, raddi, width=width, bottom=0.0)

# 颜色设置
for r, bar in zip(raddi, bars):
    bar.set_facecolor(plt.cm.viridis(r / 10.))
    bar.set_alpha(0.5)
plt.show()
```

```python
# 直方图
np.random.seed(0)
mu, sigma = 100, 20
a = np.random.normal(mu, sigma, size=100)
# 20表示直方图的个数
a = plt.hist(a, 50, histtype='step',facecolor='b', alpha=0.65)
plt.show()
```
